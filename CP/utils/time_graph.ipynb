```python
import matplotlib.pyplot as plt
rotation = []
rotation_mean = []
no_rotation = []
no_rotation_mean = []
with open('Chuffed,first_fail,im,free_search.txt','r') as f:
    for i,line in enumerate(f.readlines()):
        no_rotation.append(float(line.split(' ')[1]))
        if float(line.split(' ')[1]) != 0 and i+1!=19 and i+1!=26 and i+1!=33:
            no_rotation_mean.append(float(line.split(' ')[1]))
with open('Chuffed,input_order,im,free_search.txt','r') as f:
    for i,line in enumerate(f.readlines()):
        rotation.append(float(line.split(' ')[1]))
        #if i+1 != 11 and i+1!=23 and i+1!=26 and i+1!=27 and i+1!=33 and i+1!=35:
        if float(line.split(' ')[1]) != 0 and i+1!=17 and i+1!=24:
                rotation_mean.append(float(line.split(' ')[1]))

print('Chuffed,first_fail,im,free_search')
print(sum(no_rotation_mean)/39)
print('Chuffed,input_order,im,free_search')
print(sum(rotation_mean)/39)
plt.figure(figsize=(12,4))
plt.bar(range(1,len(no_rotation)+1), no_rotation,color = 'darkgreen', label='Chuffed,first_fail,im,free_search',align='edge',width=0.4)
plt.bar(range(1,len(rotation)+1), rotation,color = 'darkblue', label='Chuffed,input_order,im,free_search',align='edge',width=-0.4)
plt.xticks(range(1,len(no_rotation)+1))
plt.yscale('log')
plt.ylabel('Time (seconds)')
plt.xlabel('Instance number')
plt.grid(which='major',axis='y')
plt.legend()
plt.show()
```
